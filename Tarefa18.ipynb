```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
alfa = 1 # cmÂ²/s
deltat = 5 # s
deltax = 5 # cm
L = 50 # cm
tmax = 500 + (9)*20 # s
f0 = alfa*deltat/(deltax**2) # 
```

```python
x = np.linspace(0, L, int(L/deltax))
t = np.linspace(0, tmax, int(tmax/deltat))
```

```python
matrix = np.zeros((len(t),len(x)))
```

```python
for i,v in enumerate(matrix[0]):
    if((i != 0) and (i != len(matrix[0])-1) ):
        matrix[0][i] = 20
```

```python
copia = matrix.copy()
# print(matrix)
for i,linha in enumerate(copia):
    for idx,elem in enumerate(linha):
        if(i > 0):
            if((idx != 0) and (idx != len(linha) - 1)):
                copia[i][idx] = copia[i-1][idx] + f0*(copia[i-1][idx+1] - 2*copia[i-1][idx] + copia[i-1][idx-1])    
```

```python
listaplot = []
listaplot.append(copia[0])
for i, linha in enumerate(copia):
    if(i%10==0):
        listaplot.append(linha)
        
        
for linha in listaplot:
    plt.plot(x,linha, '-')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

### Participantes: Evandro Romeiro e Rodrigo Jesus

```python
c = 0.03 #30 mm
h = 15 # W/m²K
k = 200 #W/mK
P = 4*c
A = c**2
L = 0.3 #300 mm
m = math.sqrt(h*P/(k*A))
T0 = 300 #K temperatura ambiente
Tf = 350 #K temperatura da aleta
```

```python
def equacao(x):
    return (math.cosh(m*(L - x)) + h/(m*k)*math.sinh(m*(L - x)))/(math.cosh(m*L) + h/(m*k)*math.sinh(m*L))
```

```python
x = np.linspace(0, L, 60)
y = [T0 + (Tf-T0)*equacao(dx) for dx in x]
```

```python
plt.plot(x,y,'r-')
plt.title("Temperatura por posição")
plt.xlabel("Posição (m)")
plt.ylabel("Temperatura (K)")
plt.grid()
plt.show()
```
